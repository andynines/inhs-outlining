# Demo

These examples show how to view outputs generated by `inhs_outlining` using the `Fish` class.

### View a single example instance of a species

```python
from inhs_outlining import Fish

one_boops = Fish.example_of("Notropis", "Boops")

one_boops.show_ax()
one_boops.show_outline()
one_boops.show_reconstruction()
```

### View a single instance by ID number.

```python
fish_5099 = Fish.with_id("5099")

fish_5099.show_ax()
fish_5099.show_outline()
fish_5099.show_reconstruction()
```

### View all instances of a species.

```python
all_cyanelluses = Fish.all_of_species("Lepomis", "Cyanellus")

for fish in all_cyanelluses[:3]: # Truncated to first 3.
    fish.show_ax()
    fish.show_outline()
    fish.show_reconstruction()
```
