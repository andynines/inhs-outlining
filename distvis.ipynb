```python
def plot_dist_of(mat, index, bins=200):
    plt.title("Index " + str(index))
    x = []
    for freq, bin in zip(*np.histogram(mat[:, index], bins=bins)):
        x += ([bin] * freq)
    plt.hist(x, bins=bins)
    plt.show()
```

```python
plot_dist_of(X, 10)
```
